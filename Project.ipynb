# Project

In groups, work on the dynamic analysis of an offshore wind tubine following Tutorial 1, Exercise 1, Tutorial 2 and Tutorial 3.

The solution has to include:
- Dynamic beam
- Point mass at the nacelle
- Dynamic loads defined by wind and wave

The solution can include:
- Structural damping
- Aerodynamic damping
- Elastic stiffnes in soil foundation
- Effect of axial compression
- Irregular waves

Analysis:
- Where are the maximum moments located?
- What are the maximum moments?
- What are the natural frequencies?
- What is the response spectra for irregular waves?

Advanced project:
- Floating wind with mooring system
